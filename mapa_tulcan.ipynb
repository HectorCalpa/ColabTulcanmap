```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
##url='https://raw.githubusercontent.com/juaneladio/peru-geojson/master/peru_departamental_simple.geojson'
#url='https://raw.githubusercontent.com/yandunmarco/mapa_tulcan/main/yandunmarco-map_Tulcan_DOS.geojson'
url='https://raw.githubusercontent.com/yandunmarco/mapa_tulcan/main/Centros_vacunacion_Carchi2.geojson'
region_geojson = gpd.read_file(url)
region_geojson.head(31)
```

```python
region_geojson.plot(figsize=(20,20), edgecolor='black')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.show()
```

```python
regiones = list(region_geojson.ZONACIUDAD.values)
print(regiones)
```

```python
#lista de adyacencia de regiones en orden alfabtico
grafo_regional = [[1,2],[0,2],[0,1,3,4],
[2,4],[2,3]]
for u in range(len(grafo_regional)):
print(regiones[u],[regiones[v] for v in grafo_regional[u]])
```

```python
from ortools.sat.python import cp_model
#crear CSP
model = cp_model.CpModel()
#variables y dominios
color = [model.NewIntVar(1,25,'color_region_'+str(i)) for i in range(len(regiones))]
print(color)
#restricciones
for u in range(len(grafo_regional)): #para cada regin u
for v in grafo_regional[u]: #para cada vecina de u
model.Add(color[u] != color[v]) #deben tener colores distintos
#solver
solver = cp_model.CpSolver()
status = solver.Solve(model)
c = []
if status == cp_model.OPTIMAL: #si se encuentra una solucin
for u in range(len(grafo_regional)): #para cada regin
print('Regin {} ({}) de color {}'.format(u,regiones[u],solver.Value(color[u])))
c += [solver.Value(color[u])]
[color_region_0(1..25), color_region_1(1..25), color_region_2(1..25), color_region_3(1..25), color_region_4(1..25), color_region_5(1..25), color_region_6(1..25), color_regio
Regin 0 (Centro_Vacunacion_Tulcan_Norte_uno) de color 3
Regin 1 (Centro_Vacunacion_Tulcan_Norte_dos) de color 1
Regin 2 (Centro_Vacunacion_Tulcan_Centro) de color 2
Regin 3 (Centro_Vacunacion_Tulcan_Sur_uno) de color 3
Regin 4 (Centro_Vacunacion_Tulcan_Sur_dos) de color 4
```
